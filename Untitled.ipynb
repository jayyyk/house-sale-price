```python
#Model Building
train2 = pd.read_csv("train_modified.csv")
test2 = pd.read_csv("train_modified.csv")
```

```python
y_train = train2.SalePrice
X_train = train2.drop(['SalePrice', 'ScreenPorch', '3SsnPorch', 'BsmtFinSF2', 'BsmtFullBath', 'BsmtHalfBath', 'source', 'HalfBath', 'LowQualFinSF'], 1, inplace=False)
```

```python
#linear Regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```
